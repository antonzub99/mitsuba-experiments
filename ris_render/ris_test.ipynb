```python
import mitsuba as mi
import drjit as dr

import os
import numpy as np
from matplotlib import pyplot as plt
import time
print(mi.variants())
```

```python
mi.set_variant('cuda_ad_rgb')

from custom_direct_new import RISIntegrator, ISIRIntergrator

mi.register_integrator("ris_direct", lambda props: RISIntegrator(props))
```

```python
simple_box = mi.cornell_box()

simple_box['integrator'] = {
        'type': 'ris_direct',
        'emitter_samples': 32,
        'bsdf_samples': 1,
        'check_visibility': True
}

scene_ris = mi.load_dict(simple_box)

start = time.perf_counter()
img_ris = mi.render(scene_ris, spp=10)
elapsed_ris = time.perf_counter() - start
print(elapsed_ris)
```

```python
mi.Bitmap(img_ris)
```
